```python
frames = [split1_updated, split2_updated, split3_updated, split4_updated]
result = pd.concat(frames)
```

```python
label_encoder = LabelEncoder()
for col in result.columns[result.dtypes == "object"]:
    result[col] = label_encoder.fit_transform(result[col].astype('str'))
result
```

```python
X = result.drop('target', axis=1)
y = result['target']
```

```python
X_train, X_true, y_train, y_true = train_test_split(X, y, random_state=0)
```
